## 1.ราคาเฉลี่ยของบ้านที่มีราคาระหว่าง 3-5 ล้านบาท

```python
data1[(data1['price_min']>=3000000) & (data1['price_min'] <=5000000 ) ].describe() # หาบ้านที่ราคาระหว่าง 3-5 ล้านบาท
```

```python
m  = data1[ (data1['price_min']>=3000000) & (data1['price_min'] <=5000000 ) ].describe() # นำข้อมูลราคาบ้านเข้าตาราง m
me = m.iloc[1,4] # เลือก column และ row ที่สนใจ (mean) และใส่ตัวแปรเป็น me 
print(f'ราคาเฉลี่ยของบ้านที่มีราคาระหว่าง 3-5 ล้านบ้าน ประมาณ {me:,.2f} บาท' )
```

## 2.พื้นที่ของ และ จำนวนห้องนอนของบ้านที่แพงที่สุด

```python
data1['price_min'].max()  # หาบ้านที่ราคาแพงที่สุด
```

```python
P = data1[data1['price_min']==580000000.0] # นำข้อมูลของบ้านที่แพงที่สุดเข้าตาราง P
P
```

```python
am = P['area_usable_min'][33122] # หาจำนวนพื้นที่ของบ้านที่แพงที่สุด โดยใช้การชี้ค่าในตาราง
cb = P['count_room_bed'][33122]  # หาจำนวนห้องนอนของบ้านที่แพงที่สุด โดยใช้การชี้ค่าในตาราง
print(f'พื้นที่สอยของบ้านที่แพงที่สุดมีพื้นที่ประมาณ {am:,} ตารางเมตร และจำนวนห้องนอนของบ้านที่แพงที่สุดมีจำนวน {cb:.0f} ห้อง')
```
